```python
datos = pd.read_csv("casasboston.csv")
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

# Grafico

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
#Frecuncia en el numero de habitaciones
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1)
plt.show()
```



# Graficos corregidos

```python
df['N_HABITACIONES_MEDIO'].plot.hist(color='purple', edgecolor='black')
plt.title('Histograma del Nmero Medio de Habitaciones')
plt.xlabel('Nmero Medio de Habitaciones')
plt.ylabel('Frecuencia')
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=0.5, color='purple')
plt.xlabel('Valor Mediano')
plt.ylabel('Indice de crimen')
plt.title('Relacin entre valor mediano y ndice de crimen')
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh(color='purple', edgecolor='black')

plt.title('Valor Mediano de las Viviendas por Ciudad (Top 10)')
plt.xlabel('Valor Mediano')
plt.ylabel('Ciudad')

plt.show()
```
