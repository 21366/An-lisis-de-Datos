```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

texto =  open("Nube.txt").read()
Nube = WordCloud(width = 1000,
                 height=1000).generate(texto)
plt.imshow(Nube, interpolation = 'bilinear' )

plt.axis("off")
plt.show()
```
